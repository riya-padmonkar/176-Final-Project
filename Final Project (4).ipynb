# Housing in Major Cities
## Los Angeles

```python
#importing pandas
import pandas as pd 
#importing matplotlib 
import matplotlib.pyplot as plt 
```

```python
#reading in the California Houses dataset
df_la = pd.read_csv('California_Houses.csv')
#creating a new column to determine the average number of rooms per household on each block
df_la['Room'] = df_la['Tot_Rooms'] / df_la['Households']
df_la['Bedroom'] = df_la['Tot_Bedrooms'] / df_la['Households']

def cleaning(cleandf):
    #dropping all duplicates
    cleandf = cleandf.drop_duplicates()
    #replace all NaNs with 0
    cleandf = cleandf.fillna(0)
    cleandf.reset_index(drop=True, inplace=True)
    #returing the dataset
    return cleandf
clean_df = cleaning(df_la.copy())
# filtering data set to only include houses in a 40,000 meter radius of the center of LA 
def filtering(filtereddf):

    filtereddf = filtereddf[filtereddf['Distance_to_LA'] <= 40000]
    return filtereddf

filtered_df = filtering(clean_df.copy())
display(filtered_df)
#sorting data in a ascending order of Median House Value
def sort(sorteddf):
    sorteddf = sorteddf.sort_values(by='Median_House_Value', ascending=True)
    return sorteddf
sort_df = sort(filtered_df.copy())
display(sort_df)
```

```python
#creating a scatter plot of Average Rooms and Bedrooms per households
plt.scatter(sort_df['Room'], sort_df['Bedroom'], alpha=0.6, color='blue', edgecolors='w', s=80)
plt.title('Average Rooms vs. Bedrooms per Household')
plt.xlabel('Average Number of Rooms')
plt.ylabel('Average Number of Bedrooms')
plt.tight_layout()
plt.show()
```

# Boston

```python
#importing seaborn
import seaborn as sns
#reading in Boston.csv file
boston_df = pd.read_csv('Boston.csv')
boston_df.head()
display(boston_df)
```

```python
#cleaning the Boston dataset
boston_df = boston_df.drop_duplicates()
boston_df = boston_df.dropna()
```

```python
features = ['LSTAT', 'NX', 'RM']

fig, axes = plt.subplots(1, 3, figsize=(18, 6))  # Adjust size as needed

for i, feature in enumerate(features):
    #creating scatterplot
   sns.scatterplot(data=boston_df, x=feature, y='MEDV', ax=axes[i])  # Corrected: sns.scatterplot
   axes[i].set_title(f'{feature} vs MEDV')
    #setting x, y labels 
   axes[i].set_xlabel(feature)
   axes[i].set_ylabel('MEDV')

plt.tight_layout()
plt.show()
```
